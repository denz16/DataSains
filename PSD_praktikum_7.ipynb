```python
import pandas as pd
import numpy as np
```

```python
n_rows = 5
n_cols = 2
cols = ['omset', 'operasional']

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)), columns=cols)
df
```

```python
#underscore dipakai untuk ignore character
#omset konversi/kalikan ke dalam satuan ratusan ribu, operasional persepuluh ribu
df['omset'] = df['omset'] *100_000
df['operasional'] = df['operasional'] *100_000
df
```

```python
#membentuk kolom tanggal, generate index menggunakan date time
#frequencynya daily/harian

df.index = pd.util.testing.makeDateIndex(n_rows, freq='D')
df = df.reset_index()
df = df.rename(columns={'index':'tanggal'})
df
```

```python
#memasukkan formatting tampilan data frame
#sesuaikan dengan variable data frame, urutan tidak mempengaruhi, key harus sesuai nama kolom
#formatting disampaikan dalam tipe data string, maka diapit oleh tanda kutip
#.27 menyertakan dua digit extra dibelakang koma

formatku = {'tanggal':'{:%d/%m/%y}', 'operasional':'Rp {:.2f}', 'omset':'Rp {:.2f}'}
laporan = df.style.format(formatku)
laporan
```

```python
type(laporan) #tipe data (objek styler)
```

```python
#formating tanpa index
laporan.hide_index()
```

```python
#menyertakan judul dari frame
laporan.set_caption('Data Omset dan Operasional')
```

```python
laporan.highlight_min('omset', color='orange')
laporan.highlight_max('omset', color='red')

laporan.highlight_min('operasional', color='black')
laporan.highlight_max('operasional', color='green')
```

```python
#Menggabungkan (merge) dua Data Frame secara berdampingan
d1 = {'col1':[1, 2, 3], 'col2':[10, 20, 30]}
df1 = pd.DataFrame(d1)
df1
```

```python
d2 = {'col1':[4, 5, 6], 'col2':[40, 50, 60]}
df2 = pd.DataFrame(d2)
df2
```

```python
#merging yang dilakukan sifatnya ke samping, jadi ada 4 kolom

df3 = pd.merge(df1, df2, left_index=True, right_index=True) #parameter left & right
df3
```

```python
data_baru = pd.read_csv('/content/Iris.csv')
data_baru
```

```python
#mengenal groupby() dan fungsi agregasi

data_baru.groupby('Species')['PetalLengthCm'].count().to_frame()
#dikelompokkan berdasarkan species , dikolom petallenght dikenakan fungsi agregasi 
#count = hitung jumlah baris 
#to_frame = menjadikan frame baru
```

```python
data_baru.groupby('Species')['PetalLengthCm'].median().to_frame()
#setelah dikelompokkan, setiap data akan dihitung mediannya
```

```python
data_baru.groupby('Species')['PetalLengthCm'].agg(['count', 'mean', 'median'])
```

```python
data_baru.groupby('Species')['PetalLengthCm'].describe()
#describe = menampilkan deskriptif statistik
```
