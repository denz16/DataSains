# menghapus (drop) missing values (NaN)

```python
import pandas as pd

print(pd.__version__)
```

```python
df = pd.util.testing.makeMissingDataframe().reset_index()
df.head()
```

```python
df = df.rename(columns={'index':'nama'})
df.head()
```

```python
df_backup  = df.copy(deep=True)
```

```python
#menghapus (drop) setiap kolom yg mengandung missing values

df = df.dropna(axis='columns')
df.head()
```

```python
# menghapus (drop) setiap baris yang mengandung missing values

df = df_backup.copy(deep=True)
df = df.dropna(axis='rows')
df.head()
```

```python
#presentase missing values untuk setiap kolom

df = df_backup.copy(deep=True)
df.isna().mean()
```

```python
#menghapus (drop) setiap kolom yang mengandung missing values berdasarkan treshold

treshold = len(df) * 0.9
df = df.dropna(thresh=treshold, axis='columns')
df.head()
```
