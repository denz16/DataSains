```python
import numpy as np
import pandas as pd
```

```python
n_rows = 10
n_cols = 5
cols = tuple ('ABCDE')

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)), columns=cols)
df
```

```python
#seleksi column dan baris menggunakan loc
df.loc[[0,3,4], ['B', 'E']]
```

```python
#seleksi baris dengan kondisi
df.loc[df['B']>10, ['B', 'D', 'E']]
```

```python
#slicing data frame, start index end index inklusif
df.loc[0:4, 'B':'D']
```

```python
#data frame baru, tipe data integer
data = {'day':[1,2,10,25,12], 'month':[1,2,4,5,6], 'year':[2000,2001,2010,2015,2020]}
df = pd.DataFrame(data)
df
```

```python
df['penanggalan'] = pd.to_datetime(df[['day','month','year']])
df
```

```python
n_rows = 10
n_cols = 1
cols = ('usia',)

df_usia = pd.DataFrame(np.random.randint(1,99, size=(n_rows, n_cols)), columns=cols)
df_usia
```

```python
#pengelompokkan nilai numerik kedalam beberapa kategori menggunakan cut()
df_usia['kelompok_usia'] = pd.cut(df_usia['usia'], bins = [0, 18, 65, 99], labels = ['anak', 'dewasa', 'manula'])
df_usia
```

```python
n_rows = 5
n_cols = 5
cols = tuple ('ABCDE')

df_a = pd.DataFrame(np.random.randint(1,20, size=(n_rows, n_cols)), columns=cols)
df_a
```

```python
#pecah data frame menjadi dua
df1 = df_a.copy(deep=True)
df1 = df1.drop([1,4])
df1
```

```python
df2 = df_a.copy(deep=True)
df2 = df2.drop([0,3])
df2
```

```python
#menggunakan method merge untuk menggabungkan 2 df 
#inner dikedua dataframe datanya ada semua
#index reset kembali jadi 0

df_inner = pd.merge(df1,df2,how='inner')
df_inner
```

```python
df_outer = pd.merge(df1,df2,how='outer')
df_outer
```
