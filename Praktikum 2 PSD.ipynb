```python
import numpy as np
import pandas as pd
```

```python
#konversi tipe data string ke numerik 

data = {'col 1': ['1','2','3', 'text'],
        'col 2': ['1','2','3', '4']}

df_lain = pd.DataFrame(data)
df_lain
```

```python
#mengecek tipe data dari data lain

df_lain.dtypes
```

```python
#koversi tipe data dengan fungsi astype

df_x = df_lain.astype({'col 2': 'int'})
df_x
```

```python
#mengecektipe data dari data X

df_x.dtypes
```

```python
#konversi tipe data string ke numerik

df_lain.apply(pd.to_numeric, errors='coerce')

#coerce, jika ada yang tidak bisa dikonversi ke numerik, maka data tersebut akan digantikan menjadi NaN
```

```python
#membuat tanggal dengan random

#rumus tabel lama
n_rows = 5
n_cols = 5
cols = tuple('ABCDE')

df = pd.DataFrame(np.random.randint(1,10, size=(n_rows, n_cols)), columns=cols)

#memasukkan tanggal dengan random dari tabel lama
df.index = pd.util.testing.makeDateIndex(n_rows, freq='H')
df = df.reset_index ()

df['text'] = list('ABCDE')
df
```

```python
#memilih kolom bertipe data spesifik (untuk data number)

df.select_dtypes(include='number')
```

```python
#memilih kolom bertipe data spesifik (untuk data float)

df.select_dtypes(include='float')
```

```python
#membalik urutan kolom

df.loc[:, ::-1]
```

```python
#membalik urutan baris

df.loc[::-1, ]
```
